# Importing Libraries

```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pylab as plt
from IPython.display import display
```

# Reading and copying the data

```python
data = pd.read_excel('Data_Survey_Monkey_Output.xlsx')
display(data)
```

```python
data_modified = data.copy()
data_modified
```

# Data Cleaning

```python
id = list(data_modified.columns[0:7])
id 
```

```python
value = list(data_modified.columns[7:])
value
```

```python
data_melted = data_modified.melt(id_vars = id , value_vars = value , var_name = 'Question - Subquestion' , value_name='Answer' )
data_melted
```

# Creating column for total respondants for each question

```python
Total_Respondants = data_melted.groupby('Question - Subquestion')['Answer'].count()
Total_Respondants
```

```python
df = pd.DataFrame(data=Total_Respondants)
df
```

# Merging the new column for the dataframe

```python
data_merged = pd.merge(left = data_melted ,right = df , how = 'left' , on = 'Question - Subquestion') 
data_merged
```

```python
data_merged.rename(columns={'Answer_x': 'Answer', 'Answer_y': 'Total Respondants'}, inplace=True)
display(data_merged)
```

# Removing Nulls

```python
data_merged.isnull().sum()
```

```python
data_cleaned = data_merged.dropna(subset=['Answer']).reset_index()
data_cleaned
```

```python
data_cleaned = data_cleaned.drop('index',axis = 1)
data_cleaned
```

# Creating new column of the no. of employees with the same answer for a certain question

```python
same_answer = data_cleaned.groupby(['Question - Subquestion','Answer']).count().reset_index()
same_answer
```

```python
same_answer.rename(columns={'Total Respondants': 'Same Answers'}, inplace=True)
display(same_answer)
```

```python
same_answer = same_answer.drop(same_answer.columns[2:9] , axis=1)
same_answer
```

```python
data_final = pd.merge(left = data_cleaned ,right = same_answer , how = 'right' , on = ['Question - Subquestion' , 'Answer']) 
data_final
```

# Output as CSV file

```python
data_final.to_csv('Data_Survey_Monkey_Final.csv',header=True ,index=False)
```
